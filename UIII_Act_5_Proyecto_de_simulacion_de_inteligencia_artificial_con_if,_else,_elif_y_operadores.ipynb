```python
for i in range(2):
    # Sistema Experto de Deporte (IA basada en Reglas)

    print("====================================================")
    print("🤖 BIENVENIDO A LA IA DEPORTIVA v1.0 🤖")
    print("Contesta las preguntas para descubrir tu deporte ideal.")
    print("====================================================\n")

    # --- PASO 1: Captura de datos (Interactividad) ---
    print("Por favor, responde con 'si' o 'no' (en minúsculas).")
    es_en_equipo = input("¿Te gusta practicar deportes en equipo? (si/no): ")
    usa_balon = input("¿El deporte requiere obligatoriamente un balón o pelota? (si/no): ")
    es_al_aire_libre = input("¿Prefieres que sea al aire libre? (si/no): ")
    requiere_contacto = input("¿Te gusta que haya contacto físico directo con el rival? (si/no): ")

    print("\n----------------------------------------------------")
    print("🧠 PROCESANDO TUS RESPUESTAS CON NUESTRO ALGORITMO...")
    print("----------------------------------------------------\n")

    # --- PASO 2: Árbol de decisión (IA con condicionales y operadores) ---

    # RAMA 1: Deportes en equipo y con balón
    if es_en_equipo == "si" and usa_balon == "si":
        if es_al_aire_libre == "si" and requiere_contacto == "si":
            print("⚽ ¡Tu deporte ideal es el FÚTBOL o el RUGBY!")
            print("Regla activada: Equipo + Balón + Aire Libre + Contacto.")

        elif es_al_aire_libre == "si" and requiere_contacto == "no":
            print("⚾ ¡Tu deporte ideal es el BÉISBOL!")
            print("Regla activada: Equipo + Balón + Aire Libre + Sin Contacto.")

        elif es_al_aire_libre == "no" and requiere_contacto == "no":
            print("🏀 ¡Tu deporte ideal es el BALONCESTO o el VOLEIBOL de sala!")
            print("Regla activada: Equipo + Balón + Techado + Sin Contacto.")

        else:
            print("🤽 ¡Tu deporte ideal podría ser el WATERPOLO o el BALONMANO!")
            print("Regla activada: Equipo + Balón + Combinación alternativa.")

    # RAMA 2: Deportes en equipo pero SIN balón
    elif es_en_equipo == "si" and usa_balon == "no":
        if es_al_aire_libre == "si":
            print("🚣 ¡Tu deporte ideal es el REMO en equipo o las CARRERAS DE RELEVOS!")
            print("Regla activada: Equipo + Sin Balón + Aire Libre.")
        else:
            print("🤸 ¡Tu deporte ideal es la GIMNASIA RÍTMICA EN GRUPO o NATACIÓN SINCRO!")
            print("Regla activada: Equipo + Sin Balón + Techado.")

    # RAMA 3: Deportes individuales CON balón
    elif es_en_equipo == "no" and usa_balon == "si":
        if requiere_contacto == "no" and es_al_aire_libre == "si":
            print("🎾 ¡Tu deporte ideal es el TENIS o el GOLF!")
            print("Regla activada: Individual + Balón + Sin Contacto + Aire Libre.")
        else:
            print("🏓 ¡Tu deporte ideal es el TENIS DE MESA o el BILLAR!")
            print("Regla activada: Individual + Balón + Techado.")

    # RAMA 4: Deportes individuales y SIN balón (Gimnasio, combate, atletismo)
    elif es_en_equipo == "no" and usa_balon == "no":
        if requiere_contacto == "si":
            print("🥊 ¡Tu deporte ideal son las ARTES MARCIALES (Boxeo, Judo, Karate)!")
            print("Regla activada: Individual + Sin Balón + Contacto.")

        elif es_al_aire_libre == "si" and requiere_contacto == "no":
            print("🏃 ¡Tu deporte ideal es el ATLETISMO, CICLISMO o la NATACIÓN!")
            print("Regla activada: Individual + Sin Balón + Aire Libre + Sin Contacto.")

        else:
            print("🏋️ ¡Tu deporte ideal es la LEVANTAMIENTO DE PESAS o la GIMNASIA!")
            print("Regla activada: Individual + Sin Balón + Techado + Sin Contacto.")

    else:
        print("❌ Error en la matriz de datos.")
        print("Por favor, asegúrate de escribir exactamente 'si' o 'no' en minúsculas.")

    print("\n====================================================")
    print("Fin de la simulación. ¡Gracias por usar la IA Deportiva!")
    print("programa creado por Daniel Nieves NL=45")
    print("====================================================")
```
